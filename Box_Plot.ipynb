```python
#food
m = pd.read_csv('food similarity - Moinul.csv')
n = pd.read_csv('food similarity - Nafis.csv')
a = pd.read_csv('food similarity - Anika.csv')
count=0
for i in m['Value']:
    if type(i)!=float:
        count+=1
similar_count=0
for i in range(count):

    if m['Similarity'][i]!=1:
        m['Similarity'][i]=0
    if n['Similarity'][i]!=1:
        n['Similarity'][i]=0
    if a['Similarity'][i]!='1':
        a['Similarity'][i]=0
for i in range(count):
    if int(m['Similarity'][i])==int(n['Similarity'][i])==int(a['Similarity'][i]):
        similar_count+=1
print(similar_count/count)
```
